**Introduction**

The purpose of this data is to identify the relationship between education spending and various characteristics of a country, such as gdp per capita, happiness, homicide rate, and life expectancy. Our research question is:
**Does the percent of GDP a country spends on education have an impact on happiness, homicide rate, and life expectancy, and is that relationship independent from GDP per capita?**

This is an important question to ask because an increased GDP per capita is obviously going to improve issues like happiness, life expectancy, and homicide rate in a country. However, GDP per capita is not an easy metric for a country to change. If we can prove that the percent of GDP a country spends on education can improve happiness, life expectancy, and homicide rates independently from GDP per capita, then we could provide a more realistic option for governments to improve quality of life for their citizens.


Through the analysis of our data, we were able to identify many striking relationships between different variables. We found that as the percent of a country's GDP invested in education increases, so does happiness and life expectancy. One interesting finding from our data is that the percent of a country's GDP invested in education has no significant correlation with GDP per capita. We also found that both the GDP per capita and the percent of a country's GDP invested in education are positively correlated with happiness and life expectancy. This means that GDP per capita and the percent of GDP spent on education could improve happiness and life expectancy independently from one another. 


There are many incredibly impoverished countries in the world where living conditions are poor and the future is not bright. This data is important because it offers a method that could improve living conditions in these countries. According to the findings from this data, the more a country spends on education (compared to their GDP), the happier their citizens could be and the longer they might live. While these issues are, of course, much more complicated than this data might make them seem, the point of the dataset is to highlight the importance of education. A country cannot dramatically increase spending in education easily, but the importance of education for the long term quality of life for the average citizen is often undervalued. This data provides empirical evidence that education could be key to improving conditions for the average person in a developing country, which could hopefully be considered when these countries have room for development.

**Data Analysis and Evaluation of Significance**

The first and one of the most interesting findings from this dataset is the relationship between GDP per capita and the percent of GDP spent on education. We found that these two variables have a negative correlation. On average, for every 1000 dollar increase in GDP per capita, the percent of GDP spent on education decreases by about 0.0019. The linear correlation for this relationship is about -.0244. Based on our hypothesis tests, this relationship could have been produced from random noise. So, this data probably has no significant relationship. Most of the points on this graph are clustered in the bottom left section, having a low GDP per capita and investing somewhere between 0 and 6 percent of their GDP on education. Some of the notable points on this graph, however, are Cuba and American Samoa. Both of these countries have a low GDP per capita, 3.8 and 11.5 thousand (respectively), but they also have a spend a high percent of their total GDP on education, 12.2 and 14.6 (respectively).

The next relationships we investigated were homicide rate given percent GDP spent on education and homicide rate given GDP per capita. We found that as percent of GDP spent on education increased by 1, homicides per 1000 people increased by about 0.4. This data did not have a strong relationship, with a correlation of 0.05. Based on our hypothesis tests, this relationship could be a result of noise. This data was surprising both because of the positive slope, and because we expected there to be a greater correlation between the two variables. The relationship between GDP per capita and homicide rate, on the other hand, is strongly negative, with a slope of -0.13 and a rank correlation of -0.51. This is a significant value based on our hypothesis testing, since the rank correlation is greater than .3. Based on this data, it is likely that percent GDP spent on education does not have much of an effect on homicide rate and other factors are affecting this variable.

Next, we investigated happiness rankings given GDP per capita and the percent of GDP spent on education. We found that, not surprisingly, GDP per capita and happiness are positively correlated with a rank correlation of 0.83 and a slope of 0.038. This is most likely a significant value since it exceeds all values generated in our hypothesis test. In addition, this means that for every 1000 dollar increase in GDP, the happiness ranking increases by about 0.038. This is a relatively significant increase for an index that is on a scale from 1 to 8. However, percent of GDP spent on education also has a positive relationship with happiness rankings. According to our graph, when the percent of GDP spent on education increases by 1 percent, the happiness score increases by about 0.28. This relationship has a rank correlation of about 0.37, which is a relatively strong correlation that is unlikely to be created from random data according to our hypothesis tests. An interesting point to be made about this data is that since both GDP per capita and the percent of GDP invested in education are both positively correlated with happiness but have no significant relationship with each other, it is unlikely that GDP per capita is the underlying cause of increased happiness as percent GDP invested in education increases. There may be other factors at play, but GDP per capita is unlikely to be one of them.

Finally, we investigated life expectancy in these countries. We found that as GDP per capita increased by 1000, life expectancy increased by about 0.24 years on average. This relationship was strong, with a rank correlation of 0.89, which is significant based on our hypothesis tests. We also found that as percent of GDP spent on education increased by 1 percent, life expectancy increased by 1.5 years. However, this relationship was less strong, and had a rank correlation of 0.3. Despite a weaker correlation, this value is not insignficant based on our hypothesis testing, and such a relationship is unlikely to be produced from random data. The positive relationship between GDP per capita and life expectancy was not particularly surprising, since countries with a wealthier population will have access to better medical resources. However, the relationship between percent of GDP spent on education and life expectancy was more surprising. Even though percent GDP spent on education and GDP per capita do not have a significant relationship, percent GDP spent on education and life expectancy have a positive relationship. This finding helps support the argument that if more resources are invested in education, the country's population will have an improved quality of life.

```python
#Read CSV with data to be evaluated
gdp_data= pd.read_csv('final_data.csv', index_col='Country')
```

**Percent GDP invested in education given GDP per capita**

```python
#Create graph for percent GDP invested in education given GDP per capita
pyplot.scatter(gdp_data["GDP per capita (thousands)"], gdp_data["Percent GDP invested in education"])
pyplot.xlabel('GDP per Capita (thousands)')
pyplot.ylabel('Percent GDP invested in education')
pyplot.show()

#Generate mean and standard deviation for both variables
print('GDP mean = ' + str(np.mean(gdp_data['GDP per capita (thousands)'])))
print('GDP standard deviation = ' + str(np.std(gdp_data['GDP per capita (thousands)'])))
print('Percent GDP invested in education mean = ' + str(np.mean(gdp_data['Percent GDP invested in education'])))
print('Percent GDP invested in education standard deviation = ' + str(np.std(gdp_data['Percent GDP invested in education'])) + "\n")

#Create linear regression model and generate slope, intercept, and correlation for the model
gdp_education_model = LinearRegression().fit(gdp_data[['GDP per capita (thousands)']], gdp_data['Percent GDP invested in education'])
print("The regression slope for the percent of GDP invested in education given the GDP per capita = " + str(gdp_education_model.coef_[0]))
print("The intercept for the percent of GDP invested in education given the GDP per capita = " + str(gdp_education_model.intercept_))
print("Correlation for percent GDP invested in education given the GDP per capita = " + str(np.corrcoef(gdp_data["GDP per capita (thousands)"], gdp_data["Percent GDP invested in education"])[0][1]))
```

```python
#Function that creates permuted y values for GDP per capita
def permuted_y():
    return np.random.permutation(gdp_data["GDP per capita (thousands)"].copy().rank())

#Create permuted slopes for GDP per capita
permuted_slopes = []
for yy in range(1000):
    h =permuted_y()  
    
    x_permutedy_corrcoef = np.corrcoef(gdp_data["GDP per capita (thousands)"].rank(),h)[0][1]
    permuted_slopes.append(x_permutedy_corrcoef)

#Generate graph of permuted slopes for GDP per capita
pyplot.hist(permuted_slopes,bins=30)
pyplot.xlabel('GDP slopes')
pyplot.ylabel('count')
pyplot.show()
```

```python
#Function that creates permuted y values for percent GDP invested in education
def permuted_y2():
    return np.random.permutation(gdp_data["Percent GDP invested in education"].copy())

#Create permuted slopes for percent GDP invested in education
permuted_slopes = []
for yy in range(1000):
    h =permuted_y2()  
    
    x_permutedy_corrcoef = np.corrcoef(gdp_data["Percent GDP invested in education"].rank(),h)[0][1]
    permuted_slopes.append(x_permutedy_corrcoef)

#Generate graph of permuted slopes for percent GDP invested in education
pyplot.hist(permuted_slopes,bins=30)
pyplot.xlabel('Percent GDP invested in education slopes')
pyplot.ylabel('count')
pyplot.show()

```
